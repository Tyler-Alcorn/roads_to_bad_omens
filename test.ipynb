```python
from bs4 import BeautifulSoup
import json

# Load the uploaded file
file_path = "history.html"

try:
    with open(file_path, "r", encoding="utf-8") as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, "html.parser")
    video_entries = []

    # Target the specific container identified in your file
    cards = soup.find_all("div", attrs={"jsname": "r4nke"})

    for card in cards:
        # Find the link
        link_tag = card.find("a", href=True)
        
        if link_tag and "watch?v=" in link_tag["href"]:
            # Extract the title from the span with class 'hFYxqd'
            title_span = link_tag.find("span", class_="hFYxqd")
            title = title_span.get_text(strip=True) if title_span else link_tag.get_text(strip=True)
            
            # Extract the timestamp (usually text right before the link or in the parent)
            # Based on your file, the text "Watched" and the time are siblings to the <a> tag
            timestamp = card.get_text(separator=" ", strip=True).split("Watched")[-1].strip()

            video_entries.append({
                "title": title,
                "url": link_tag["href"],
                "watched_at": timestamp
            })

    # Save to JSON
    output_file = "youtube_history_sample.json"
    with open(output_file, "w", encoding="utf-8") as out:
        json.dump(video_entries, out, indent=4, ensure_ascii=False)

    print(f"✅ Success! Extracted {len(video_entries)} videos.")
    print(f"Sample Entry: {video_entries[0]['title']} at {video_entries[0]['watched_at']}")

except Exception as e:
    print(f"❌ Error: {e}")
```

```python
from bs4 import BeautifulSoup
import json

file_path = "history.html"

try:
    with open(file_path, "r", encoding="utf-8") as f:
        html_content = f.read()

    soup = BeautifulSoup(html_content, "html.parser")
    video_entries = []

    # Targeting the activity cards from your file
    cards = soup.find_all("div", attrs={"jsname": "r4nke"})

    for card in cards:
        # 1. Get the Video Link and Title
        video_link_tag = card.find("a", href=True)
        if video_link_tag and "watch?v=" in video_link_tag["href"]:
            title_span = video_link_tag.find("span", class_="hFYxqd")
            video_title = title_span.get_text(strip=True) if title_span else video_link_tag.get_text(strip=True)
            
            # 2. Get the Creator/Channel Name
            # In your file, the channel is often the SECOND link in the card
            all_links = card.find_all("a", href=True)
            creator_name = "Unknown Creator"
            
            if len(all_links) > 1:
                # The second link is usually the channel link
                creator_name = all_links[1].get_text(strip=True)
            
            video_entries.append({
                "title": video_title,
                "url": video_link_tag["href"],
                "creator": creator_name
            })

    # Save to JSON
    output_file = "youtube_history_sample_with_creators.json"
    with open(output_file, "w", encoding="utf-8") as out:
        json.dump(video_entries, out, indent=4, ensure_ascii=False)


    print(f"✅ Extracted {len(video_entries)} videos with creator info.")

except Exception as e:
    print(f"❌ Error: {e}")
```
